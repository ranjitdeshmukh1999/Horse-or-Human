```python
X = df1.drop('salary',axis = 1)
```

```python
y = df1['salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state=0,n_jobs=-1)
```

```python
lr_fit = lr.fit(X_train,y_train)
```

```python
lr_pred = lr_fit.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
accuracy_score(lr_pred,y_test)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=0,n_jobs=-1)
```

```python
rf_fit = rf.fit(X_train, y_train)
```

```python
rf_pred = rf_fit.predict(X_test)
```

```python
accuracy_score(rf_pred,y_test)
```

```python
classification_report(rf_pred,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = { 'n_estimators' : [ 100,200,300,400,500],
           'max_depth' : [5,10,15,20,25]}
```

```python
grid = GridSearchCV(RandomForestClassifier(),params)
```

```python
grid_search = grid.fit(X_train,y_train)
```

```python
grid_search.get_params
```

```python
grid_search.best_params_
```

```python
rfgrid = RandomForestClassifier(max_depth=15,n_estimators=500,n_jobs=-1,random_state=0)
```

```python
rfgrid_fit = rfgrid.fit(X_train,y_train)
```

```python
rfgrid_pred = rfgrid_fit.predict(X_test)
```

```python
accuracy_score(rfgrid_pred,y_test)
```
